### 2. Examine the State and Action Spaces

### 4. It's Your Turn!

Now it's your turn to train your own agent to solve the environment!  When training the environment, set `train_mode=True`, so that the line for resetting the environment looks like the following:
```python
env_info = env.reset(train_mode=True)[brain_name]
```

````python
from DQNAgent import Agent
agent = Agent(state_size, action_size)
````

````python
def DQN(nEpisodes = 2000, EPSStart = 1.0, EPSEnd = 0.01, EPSDecay = 0.995):
    N = 100
    scores = []
    scoresWindow = deque(maxlen = N)
    eps = EPSStart
    for episode in range(1, nEpisodes + 1):
        env_info = env.reset(train_mode=True)[brain_name]
        state = env_info.vector_observations[0]
        score = 0
        while True:
            action = int(agent.act(state, eps))
            env_info = env.step(action)[brain_name]
            nextState = env_info.vector_observations[0]
            reward = env_info.rewards[0]
            done = env_info.local_done[0]
            agent.step(state, action, reward, nextState, done)
            state = nextState
            score += reward
            if done:
                break
        scoresWindow.append(score)
        scores.append(score)
        eps = max(EPSEnd, EPSDecay*eps)
        
        # Print scores while training
        print('\rEpisode {}\tAverage Score: {:.2f}'.format(episode, np.mean(scoresWindow)), end="")
        if episode % N == 0:
            print('\rEpisode {}\tAverage Score: {:.2f}'.format(episode, np.mean(scoresWindow)))
            
        if np.mean(scoresWindow) >= 13.:
            print('\nEpisode: {:d}'.format(episode))
            print('\nEnvironment solved in {:d} episodes!\tAverage Score: {:.2f}'.format(episode-N, np.mean(scoresWindow)))
            torch.save(agent.qnetworkLocal.state_dict(), 'modelWeights.pth')
            break;
    return scores

scores = DQN()
````

````python
# Plot the scores
fig = plt.figure(figsize=(20, 10))
ax = fig.add_subplot(111)
plt.plot(np.arange(len(scores)), scores)
plt.xlabel('Episode Number')
plt.ylabel('Score')
plt.show()
````
